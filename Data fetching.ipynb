```python
import pandas as pd
import numpy as np
import seaborn as sns
import requests
import json
# https://recruitment.aimtechnologies.co/ai-tasks
```

```python
df=pd.read_csv('dialect_dataset.csv')
df
```

```python
id=df['id'].tolist()
id = [str(x) for x in id]
```

```python
url = "https://recruitment.aimtechnologies.co/ai-tasks"
# headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}
```

```python
text={}
for i in range(0,len(id),1000):
    id_batches=id[i:i+1000]
    request = requests.post(url, data=json.dumps(id_batches))
    while(True):
        if request.status_code == 200:
            text.update(json.loads(request.text))
            print(len(text))
        break
```

```python
len(text)
```

```python
df2=pd.DataFrame(text.items(), columns=['id', 'text'])
df2
```

```python
data = pd.concat([df, df2['text']], axis=1)
```

```python
data.to_csv('dialect_dataset with text.csv', encoding='utf-8')
```
